```python
########importing libraries #################
import numpy as np
import cv2
from matplotlib import pyplot as plt
import glob
import pandas as pd
```

```python
aaimage_list = []
location_of_files='*.tif'
fileList = glob.glob(location_of_files)
dict1={}
```

```python
for i in fileList:
    #print(i),
    #print('------>')
    im = cv2.imread(i)
    imgray = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
    maxValue = 255
    thresh= 80
    th,dst = cv2.threshold(im,thresh, maxValue, cv2.THRESH_BINARY_INV)
    imgrey = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(imgrey, 127, 255, 0)
    im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(dst, contours, -1, (0,255,0), 3)
    #cv2.imwrite(i,dst)  
    
    mylist =[]
   
    for j in contours:
        area = cv2.contourArea(j)
        mylist.append(area)
        #print(area)
        #print('------------')
   # mylist.sort(reverse=True)
    print(i,'----->',len(mylist))
    dict1[i] = mylist   

 
 
```

```python
df = pd.DataFrame.from_dict(dict1,orient='index') 
```
