```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from math import sqrt
from skimage.color import rgb2gray
import glob
from skimage.io import imread
import pandas as pd
location_of_files = '*.tif'
fileList = glob.glob(location_of_files)
image_list = []
LoG_list = []
DoG_list = []
DoH_list = []
print("Image \t\t       LoG\t\tDoG\t\tDoH")
for i in fileList:
    im = imread(i, as_grey=True)
    image_list.append(i)
    
    ##########____LoG_calculation_______#############
    blobs_log = blob_log(im,min_sigma=15, max_sigma=60, threshold=.1)
    # Compute radii in the 3rd column.
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
    LoG = len(blobs_log)
    LoG_list.append(LoG)
    
    ##########____DoG_calculation_______############
    blobs_dog = blob_dog(im,min_sigma=15, max_sigma=60, threshold=.1)
    # Compute radii in the 3rd column.
    blobs_dog[:, 2] = blobs_dog[:, 2] * sqrt(2)
    DoG = len(blobs_dog)
    DoG_list.append(DoG)
    
    

    ##########____DoH_calculation_______############
    blobs_doh = blob_doh(im,min_sigma=15, max_sigma=60, threshold=.1)
    # Compute radii in the 3rd column.
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)
    DoH = len(blobs_doh)
    DoH_list.append(DoH)


    print(i,"\t|\t",LoG,"\t|\t",DoG,"\t|\t",DoH)
   
 
    ##########_________EndOfLoop___________########

#writing data to excel sheet   
df = pd.DataFrame({'DoH Value': DoH_list,
                   'DoG Value': DoG_list,
                   'LoG Value': LoG_list,
                   'name': image_list,})
writer = pd.ExcelWriter('result_insitu.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1',index=False)
writer.save()
```
