```python
########importing libraries #################
import numpy as np
import cv2
from matplotlib import pyplot as plt
import glob
import pandas as pd
```

```python
location_of_files='*/*.tif'
fileList = glob.glob(location_of_files)
imagelist=[]
countlist=[]
for i in fileList:
    im = cv2.imread(i)
    imgray = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
    maxValue = 255
    thresh= 80
    th,dst = cv2.threshold(im,thresh, maxValue, cv2.THRESH_BINARY_INV)
    imgrey = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(imgrey, 127, 255, 0)
    im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(dst, contours, -1, (0,255,0), 3)
    
    mylist =[]
    
   
    for j in contours:
        area = cv2.contourArea(j)
        mylist.append(area)
        
    print(i,'----->',len(mylist))
    imagelist.append(i)
    countlist.append(len(mylist))
    
   
```

```python
df = pd.DataFrame({' Value': countlist,
                   'name': imagelist,})
df
```

```python
writer = pd.ExcelWriter('result.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1',index=False)
writer.save()
```
